```python
# 연도별로 처리할 컬럼명 정의
col_config = {
    2022: {
        "df": df22,
        "group_col": "A7",                    # 여행 동반자 수
        "age_col": "BAGE",                    # 연령대
        "region_col": "D_TRA1_1_SPOT",        # 1번째 방문지역
        "sex_col": "BSEX",                    # 성별
        "spend_col": "D_TRA1_ONE_COST",       # 1인 지출비용
        "stay_col": "D_TRA1_1_Q6",            # 숙박 형태
        "stay_day_col": "D_TRA1_S_Day",       # 여행 일수(박)
        "total_cost_col": "D_TRA1_COST",      # 여행 총 경비
        "travel_type_col": "D_TRA1_CASE",     # 여행 유형 (관광/방문/출장 등)
        "reason1_col": "A4_1",                # 여행지 선택 이유 1순위
        "reason2_col": "A4_2",                # 여행지 선택 이유 2순위
        "reason3_col": "A4_3",                # 여행지 선택 이유 3순위
        "transport_col": "A1_1"               # 교통수단
    },
    2023: {
        "df": df23,
        "group_col": "A.[국내] Q7. 여행 동반자 수(본인포함)",
        "age_col": "공통] 연령",
        "region_col": "국내]  B_여행1차_1번째방문지_방문 지역",
        "sex_col": "공통] 성별",
        "spend_col": "국내]  B_여행1차_1인지출비용",
        "stay_col": "국내] B_여행1차_1번째방문지_숙박 시설",
        "stay_day_col": "국내] B_여행1차_여행일수(박)",
        "total_cost_col": "국내] B_여행1차_여행 총경비",
        "travel_type_col": "국내] B_여행1차_여행유형",
        "reason1_col": "A.[국내] Q4. 여행지 선택 이유_1순위",
        "reason2_col": "A.[국내] Q4. 여행지 선택 이유_2위",
        "reason3_col": "A.[국내] Q4. 여행지 선택 이유_3순위",
        "transport_col": "A.[국내] Q1. 주요 이동 교통수단_1순위"
    },
      2024: {
        "df": df24,
        "group_col": "A.[국내] Q6. 여행 동반자 수(본인 포함)",
        "age_col": "공통]  연령",
        "region_col": "국내]  B_여행1차_1번째방문지_방문 지역",
        "sex_col": "공통]  성별",
        "spend_col": "국내]  B_여행1차_1인지출비용",
        "stay_col": "국내]  B_여행1차_1번째방문지_숙박 시설",
        "stay_day_col": "국내]  B_여행1차_여행일수(박)",
        "total_cost_col": "국내] B_여행1차_여행 총경비",
        "travel_type_col": "국내] B_여행1차_여행유형",
        "reason1_col": "A.[국내] Q3. 여행지 선택 이유_1순위",
        "reason2_col": "A.[국내] Q3. 여행지 선택 이유_2순위",
        "reason3_col": "A.[국내] Q3. 여행지 선택 이유_3순위",
        "transport_col": "국내]  B_여행1차_주요 이동 교통 수단(지역간)"
    }
}
from difflib import get_close_matches

def safe_get_column(df, col_name):
    """컬럼명이 없으면 유사한 컬럼명을 반환"""
    if col_name in df.columns:
        return col_name
    else:
        match = get_close_matches(col_name, df.columns, n=1)
        if match:
            print(f"컬럼명 자동 수정: '{col_name}' → '{match[0]}'")
            return match[0]
        else:
            raise KeyError(f"컬럼 '{col_name}' not found and no close match in columns")

# 필터링 및 저장 (자동 컬럼 보정 포함)
for year, conf in col_config.items():
    df = conf["df"].copy()
    
    # 보정된 컬럼명 매핑
    corrected = {k: safe_get_column(df, v) for k, v in conf.items() if k != "df"}

    # 숫자형 변환
    df[corrected["group_col"]] = pd.to_numeric(df[corrected["group_col"]], errors="coerce")
    df[corrected["age_col"]] = pd.to_numeric(df[corrected["age_col"]], errors="coerce")
    df[corrected["spend_col"]] = pd.to_numeric(df[corrected["spend_col"]], errors="coerce")

    # 소규모 여행자 필터링 (1~4인)
    df_filtered = df[df[corrected["group_col"]].between(1, 4)].copy()

    # 필요한 컬럼만 선택
    selected_cols = [
        corrected["region_col"], corrected["age_col"], corrected["sex_col"],
        corrected["group_col"], corrected["spend_col"], corrected["stay_col"],
        corrected["stay_day_col"], corrected["total_cost_col"], corrected["travel_type_col"],
        corrected["reason1_col"], corrected["reason2_col"], corrected["reason3_col"],
        corrected["transport_col"]
    ]
    df_final = df_filtered[selected_cols].copy()

    # 컬럼명 통일
    df_final.columns = [
        "지역코드", "연령", "성별", "여행인원", "1인지출", "숙박형태", "여행일수",
        "총경비", "여행유형", "선택이유1", "선택이유2", "선택이유3", "교통수단"
    ]
    
    # 연도 컬럼 추가
    df_final["연도"] = year

    # CSV 저장
    df_final.to_csv(f"소규모여행자_{year}.csv", index=False, encoding="utf-8-sig")
    print(f"[{year}] 저장 완료: 소규모여행자_{year}.csv")
```

```python
# 대상 컬럼들
int_cols = ["지역코드", "성별", "여행인원", "여행유형", "여행일수", "연령"]

# 전처리 함수
def convert_to_int(df):
    for col in int_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(-1).astype(int)
    return df

# 적용
data22 = convert_to_int(data22)
data23 = convert_to_int(data23)
data24 = convert_to_int(data24)
```

```python
def print_stay_stats(df, year):
    # 여행일수 숫자형으로 변환
    df["여행일수"] = pd.to_numeric(df["여행일수"], errors="coerce")

    total = len(df)
    daytrip = (df["여행일수"] == 0).sum()
    stay = (df["여행일수"] >= 1).sum()
    
    print(f"[{year}] 전체 소규모 여행자 수: {total:,}명")
    print(f" - 당일치기 여행자(0박): {daytrip:,}명 ({daytrip / total * 100:.1f}%)")
    print(f" - 숙박 여행자(1박 이상): {stay:,}명 ({stay / total * 100:.1f}%)\n")

print_stay_stats(data22, 2022)
print_stay_stats(data23, 2023)
print_stay_stats(data24, 2024)
```

```python
# 선택 이유 매핑
reason_map = {
    1: "지명도", 2: "볼거리", 3: "저렴한 경비", 4: "이동거리", 5: "여행 시간",
    6: "숙박시설", 7: "쇼핑", 8: "음식", 9: "교통편", 10: "체험 프로그램",
    11: "추천", 12: "편의시설", 13: "교육성", 14: "동반자", 15: "기타"
}

# 교통수단 매핑
transport_map = {
    1: "자가용", 2: "철도", 3: "항공기", 4: "선박", 5: "지하철", 6: "정기버스",
    7: "전세버스", 8: "렌트", 9: "자전거", 10: "택시", 11: "기타"
}

# 데이터 리스트와 연도 리스트
data_list = [data22, data23, data24]
year_list = [2022, 2023, 2024]

# 시각화 함수
def process_and_plot(df, year):
    # 선택이유 매핑
    for i in range(1, 4):
        col = f"선택이유{i}"
        df[col] = pd.to_numeric(df[col], errors="coerce").map(reason_map)

    # 교통수단 매핑
    df["교통수단"] = pd.to_numeric(df["교통수단"], errors="coerce").map(transport_map)

    # 시각화
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    for i in range(1, 4):
        col = f"선택이유{i}"
        counts = df[col].value_counts().sort_values(ascending=False)
        ax = axes[(i-1)//2, (i-1)%2]
        counts.plot(kind="bar", ax=ax)
        ax.set_title(f"{year}년 선택 이유 {i}순위")
        ax.set_ylabel("응답 수")
        ax.set_xlabel("이유")
        ax.tick_params(axis='x', rotation=45)

    # 교통수단
    counts = df["교통수단"].value_counts().sort_values(ascending=False)
    ax = axes[1, 1]
    counts.plot(kind="bar", ax=ax, color="skyblue")
    ax.set_title(f"{year}년 주요 교통수단")
    ax.set_ylabel("응답 수")
    ax.set_xlabel("교통수단")
    ax.tick_params(axis='x', rotation=45)

    plt.suptitle(f"{year}년 선택 이유 및 교통수단 분포", fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# 실행
for df, year in zip(data_list, year_list):
    process_and_plot(df.copy(), year)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.cluster import KMeans

def perform_kmeans(df, n_clusters=3, random_state=42):
    df = df.copy()
    
    # 숙박여부 이진화 및 변환
    df["숙박여부"] = df["숙박여부"].map({"숙박": 1, "당일치기": 0})
    df = df[pd.to_numeric(df["숙박여부"], errors='coerce').notnull()]
    df["숙박여부"] = df["숙박여부"].astype(float)
    
    # 사용할 특성
    numeric_cols = ["1인지출", "여행일수", "여행인원", "숙박여부"]
    categorical_cols = ["숙박형태", "교통수단", "여행유형_그룹"]

    # 원핫 인코딩
    encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
    encoded = encoder.fit_transform(df[categorical_cols])
    encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(categorical_cols), index=df.index)

    # 수치형 컬럼 + 인코딩 결과 결합
    X = pd.concat([df[numeric_cols], encoded_df], axis=1)

    # 정규화
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # 시드 고정한 KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state, n_init=10)
    df["클러스터"] = kmeans.fit_predict(X_scaled)

    return df

# 각 연도별 군집분석 수행
data22_clustered = perform_kmeans(data22)
data23_clustered = perform_kmeans(data23)
data24_clustered = perform_kmeans(data24)
```

```python
# 연도별 시도별 '고지출 숙박형' 수 count
grouped_dfs = []
for year, df in zip([2022, 2023, 2024], [data22_clustered, data23_clustered, data24_clustered]):
    temp = df[df["클러스터유형"] == "고지출 숙박형"].groupby("시도명").size().reset_index(name="고지출숙박형_수")
    temp["연도"] = year
    grouped_dfs.append(temp)

# 통합
cluster_demand = pd.concat(grouped_dfs)
cluster_demand
```

```python
# 시도별 요약 지표 계산
summary_dfs = []
for df in [data22_clustered, data23_clustered, data24_clustered]:
    temp = (
        df.groupby(["연도", "시도명"])
        .agg(
            숙박률=("숙박여부", lambda x: (x == 1).mean()),
            평균지출비용=("1인지출", "mean"),
            여행자수=("지역코드", "count"),
            대표_지역코드=("지역코드", lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
        )
        .reset_index()
    )
    summary_dfs.append(temp)

summary_by_region = pd.concat(summary_dfs, ignore_index=True)

# 수요 데이터와 결합하여 infra_base 생성
infra_base = pd.merge(
    cluster_demand,
    summary_by_region,
    on=["연도", "시도명"],
    how="left"
)
```

```python
# 기준값 계산
threshold_demand = infra_base["고지출숙박형_수"].quantile(0.5)
threshold_stay = infra_base["숙박률"].median()

# 조건 추가
infra_base["수요있음"] = infra_base["고지출숙박형_수"] >= threshold_demand
infra_base["저활용여부"] = infra_base["숙박률"] < threshold_stay

# 입지 후보 필터링
infra_candidates = infra_base[
    (infra_base["수요있음"]) & (infra_base["저활용여부"])
].copy()
```

```python
# 조건 완화 설정 (수요 있고 저활용인 지역 찾기)
threshold_demand = infra_base["고지출숙박형_수"].quantile(0.4)  
threshold_stay = infra_base["숙박률"].quantile(0.6)            

infra_base["완화_저활용"] = infra_base["숙박률"] < threshold_stay
infra_base["완화_수요있음"] = infra_base["고지출숙박형_수"] >= threshold_demand

infra_candidates_relaxed = infra_base[
    (infra_base["완화_저활용"]) & (infra_base["완화_수요있음"])
].copy()

# 대표 지역코드(최빈값) 추출
region_mode_debug = pd.concat([
    data22_clustered[["시도명", "연도", "지역코드"]],
    data23_clustered[["시도명", "연도", "지역코드"]],
    data24_clustered[["시도명", "연도", "지역코드"]]
])

# 공백 제거
region_mode_debug["시도명"] = region_mode_debug["시도명"].str.strip()
infra_candidates_relaxed["시도명"] = infra_candidates_relaxed["시도명"].str.strip()

# 시도-연도별 최빈 지역코드 계산
region_mode_clean = (
    region_mode_debug
    .groupby(["시도명", "연도"])["지역코드"]
    .agg(lambda x: x.mode().iloc[0])  # 최빈값
    .reset_index()
)

# 병합
infra_candidates_relaxed = pd.merge(
    infra_candidates_relaxed,
    region_mode_clean,
    on=["시도명", "연도"],
    how="left"
)


# 최종 결과 확인
infra_candidates_relaxed[
    ["연도", "시도명", "지역코드", "고지출숙박형_수", "숙박률", "여행자수", "완화_수요있음", "완화_저활용"]
]
```

```python
# 연도별 수요 있는 입지 후보 필터링
top_candidates_by_year = (
    infra_candidates_relaxed[
        infra_candidates_relaxed["완화_수요있음"] == True
    ]
    .sort_values(by=["연도", "고지출숙박형_수"], ascending=[True, False])
    [["연도", "시도명", "지역코드", "고지출숙박형_수", "숙박률", "여행자수", "완화_수요있음", "완화_저활용"]]
)

# 결과 출력
top_candidates_by_year
```

```python
숙박형태_map = {
    1: "호텔",
    2: "모텔/여관",
    3: "펜션",
    4: "콘도미니엄/리조트",
    5: "유스호스텔",
    6: "야영(캠핑)",
    7: "연수원/수련원",
    8: "전통 한옥 숙박시설",
    9: "게스트 하우스",
    10: "민박",
    11: "자연휴양림 내 숙박시설",
    12: "가족/친지집",
    13: "무박 여행",
    14: "기타"
}
```

```python
full_data = pd.concat([data22_clustered, data23_clustered, data24_clustered], ignore_index=True)
filtered = full_data[full_data["지역코드"].isin([36020, 37020, 35010, 31370, 33380, 34030])].copy()

# 숙박형태 코드 변환 및 매핑
filtered.loc[:, "숙박형태"] = pd.to_numeric(filtered["숙박형태"], errors="coerce")
filtered.loc[:, "숙박형태명"] = filtered["숙박형태"].map(숙박형태_map)

# 집계
숙박형태_분포 = (
    filtered.dropna(subset=["숙박형태명"])
    .groupby(["연도", "시도명", "지역코드", "숙박형태명"])
    .size()
    .reset_index(name="빈도")
    .sort_values(["시도명", "연도", "빈도"], ascending=[True, True, False])
)

숙박형태_분포
```

```python
# 각 연도-지역코드 조합에서 상위 3개 숙박형태 추출
top3_by_region_year = (
    숙박형태_분포
    .sort_values(["연도", "지역코드", "빈도"], ascending=[True, True, False])
    .groupby(["연도", "지역코드"])
    .head(3)
    .reset_index(drop=True)
)

top3_by_region_year
```

```python
# 지역코드 ↔ 지역명 매핑
region_map = {
    31370: "가평", 36020: "여수", 37020: "경주",
    35010: "전주", 33380: "단양", 34030: "보령"
}
reverse_map = {v: k for k, v in region_map.items()}

# 원본 복사 및 지역명 추가
df = top3_by_region_year.copy()
df["지역명"] = df["지역코드"].map(region_map)

# 가족/친지집 제외 후 Top3 추출 함수 정의
def get_top3_excluding(group):
    제외형태 = ["가족/친지집"]
    df_filtered = group[~group["숙박형태명"].isin(제외형태)].copy()
    df_filtered = df_filtered.sort_values("빈도", ascending=False).head(3)
    df_filtered["순위"] = range(1, len(df_filtered) + 1)
    return df_filtered

# 연도-지역명 단위로 Top3 추출
group_cols = ["연도", "지역명"]
filtered_df = (
    df.set_index(group_cols)
      .groupby(level=group_cols, group_keys=False)
      .apply(get_top3_excluding)
      .reset_index()
)

# 피벗: 숙박형태명 + 빈도 분리
pivot_df = filtered_df.pivot(index=["연도", "지역명"], columns="순위", values=["숙박형태명", "빈도"])
pivot_df.columns = [f"Top{i}_{col}" for col, i in pivot_df.columns]
pivot_df = pivot_df.reset_index()

# 지역코드 복원
pivot_df["지역코드"] = pivot_df["지역명"].map(reverse_map)
```

```python
# 여행자수 및 평균 숙박일수 계산
region_summary = []
for _, row in pivot_df.iterrows():
    year = row["연도"]
    code = row["지역코드"]
    subset = full_data[
        (full_data["연도"] == year) &
        (full_data["지역코드"] == code)
    ]
    region_summary.append({
        "연도": year,
        "지역코드": code,
        "여행자수": len(subset),
        "평균숙박일수": round(subset["여행일수"].mean(), 2) if not subset.empty else 0
    })

region_summary_df = pd.DataFrame(region_summary)
region_summary_df
```

```python
# 최종 병합
final_df = pd.merge(pivot_df, region_summary_df, on=["연도", "지역코드"], how="left")

# 컬럼 정리 및 정렬
final_df = final_df[[
    "연도", "지역명", "Top1_숙박형태명", "Top1_빈도",
    "Top2_숙박형태명", "Top2_빈도",
    "여행자수", "평균숙박일수"
]].sort_values(by=["지역명", "연도"]).reset_index(drop=True)

# 결과 출력
display(final_df)
```

```python
# 연도별 나눠서 출력
for year in sorted(final_df["연도"].unique()):
    print(f"\n{year}년 지역별 숙박형태 및 여행 특성 요약")
    
    df_year = (
        final_df[final_df["연도"] == year]
        .sort_values("지역명")  # 지역 기준 정렬
        .reset_index(drop=True)
    )
    
    display(df_year)
```

```python
parking["주소"] = parking["소재지도로명주소"].fillna("")  
parking.loc[parking["주소"] == "", "주소"] = parking["소재지지번주소"]

def extract_sigungoo(address):
    try:
        parts = address.split()
        if len(parts) >= 2:
            return parts[0] + " " + parts[1]  
    except:
        return None

public = parking[parking["주차장구분"] == "공영"].copy()

public["시군구"] = public["주소"].apply(extract_sigungoo)

parking_counts = public["시군구"].value_counts().reset_index()
parking_counts.columns = ["시군구", "공영주차장_수"]

target_regions = [
    "전라남도 여수시", "경상북도 경주시", "전라북도 전주시",
    "경기도 가평군", "충청북도 단양군", "충청남도 보령시"
]

parking_counts[parking_counts["시군구"].isin(target_regions)]
```

```python
import folium
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

trip_위치["주소"] = "전라남도 여수시 " + trip_위치["위치"]

geolocator = Nominatim(user_agent="yeosu_house_locator")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

latitudes = []
longitudes = []

for addr in empty_여수_sorted["주소"]:
    try:
        location = geocode(addr)
        if location:
            latitudes.append(location.latitude)
            longitudes.append(location.longitude)
        else:
            latitudes.append(None)
            longitudes.append(None)
    except:
        latitudes.append(None)
        longitudes.append(None)

empty_여수_sorted["위도"] = latitudes
empty_여수_sorted["경도"] = longitudes

trip_lat = []
trip_lng = []

for addr in trip_위치["주소"]:
    try:
        location = geocode(addr)
        if location:
            trip_lat.append(location.latitude)
            trip_lng.append(location.longitude)
        else:
            trip_lat.append(None)
            trip_lng.append(None)
    except:
        trip_lat.append(None)
        trip_lng.append(None)

trip_위치["위도"] = trip_lat
trip_위치["경도"] = trip_lng

m = folium.Map(location=[34.7604, 127.6622], zoom_start=11)

highlight = ["돌산읍", "화양면", "동문동", "화정면", "충무동"] # 빈집 수가 많은 상위 5개 지역 보라색으로 표시

for _, row in empty_여수_sorted.dropna(subset=["위도", "경도"]).iterrows():
    eupmyeondong = row["읍면동"]
    if eupmyeondong in highlight:
        folium.Marker(
            location=[row["위도"], row["경도"]],
            popup=f"<b>{row['주소']}<br>빈집수: {row['빈집수']}</b>",
            icon=folium.Icon(color='purple', icon='flag')
        ).add_to(m)
    else:
        folium.Marker(
            location=[row["위도"], row["경도"]],
            popup=f"{row['주소']}<br>빈집수: {row['빈집수']}",
            icon=folium.Icon(color='blue', icon='home')
        ).add_to(m)

for _, row in trip_위치.dropna(subset=["위도", "경도"]).iterrows():
    folium.Marker(
        location=[row["위도"], row["경도"]],
        popup=row["인기관광지"],
        icon=folium.Icon(color='red', icon='star')
    ).add_to(m)
    
m.save("yeosu_map.html")
```
