```python
import pandas as pd

# 연도별로 처리할 컬럼명 정의 (문법/쉼표 오류 수정 + stay_price 포함)
col_config = {
    2022: {
        "df": df22,
        "group_col": "A7",                    # 여행 동반자 수
        "age_col": "BAGE",                    # 연령대
        "region_col": "D_TRA1_1_SPOT",        # 1번째 방문지역
        "sex_col": "BSEX",                    # 성별
        "spend_col": "D_TRA1_ONE_COST",       # 1인 지출비용
        "stay_col": "D_TRA1_1_Q6",            # 숙박 형태
        "stay_day_col": "D_TRA1_S_Day",       # 여행 일수(박)
        "total_cost_col": "D_TRA1_COST",      # 여행 총 경비
        "travel_type_col": "D_TRA1_CASE",     # 여행 유형 (관광/방문/출장 등)
        "reason1_col": "A4_1",                # 여행지 선택 이유 1순위
        "reason2_col": "A4_2",                # 여행지 선택 이유 2순위
        "reason3_col": "A4_3",                # 여행지 선택 이유 3순위
        "transport_col": "A1_1",              # 교통수단
        "stay_price": "A9C"                   # 숙박비 총액
    },
    2023: {
        "df": df23,
        "group_col": "A.[국내] Q7. 여행 동반자 수(본인포함)",
        "age_col": "공통] 연령",
        "region_col": "국내]  B_여행1차_1번째방문지_방문 지역",
        "sex_col": "공통] 성별",
        "spend_col": "국내]  B_여행1차_1인지출비용",
        "stay_col":       "국내]  B_여행1차_1번째방문지_숙박 시설",
        "stay_day_col":   "국내]  B_여행1차_여행일수(박)",
        "total_cost_col": "국내]  B_여행1차_여행 총경비",
        "travel_type_col":"국내]  B_여행1차_여행유형",
        "reason1_col": "A.[국내] Q4. 여행지 선택 이유_1순위",
        "reason2_col": "A.[국내] Q4. 여행지 선택 이유_2순위",
        "reason3_col": "A.[국내] Q4. 여행지 선택 이유_3순위",
        "transport_col": "A.[국내] Q1. 주요 이동 교통수단_1순위",
        "stay_price": "A.[국내] Q9. 3. 숙박비 총액[단위:원]"
    },
    2024: {
         "df": df24,
        "group_col": "A.[국내] Q6. 여행 동반자 수(본인 포함)",
        "age_col": "공통]  연령",
        "region_col": "국내]  B_여행1차_1번째방문지_방문 지역",
        "sex_col": "공통]  성별",
        "spend_col": "국내]  B_여행1차_1인지출비용",
        "stay_col": "국내]  B_여행1차_1번째방문지_숙박 시설",
        "stay_day_col": "국내]  B_여행1차_여행일수(박)",
        "total_cost_col":  "국내]  B_여행1차_여행 총경비",
        "travel_type_col": "국내]  B_여행1차_여행유형",
        "reason1_col": "A.[국내] Q3. 여행지 선택 이유_1순위",
        "reason2_col": "A.[국내] Q3. 여행지 선택 이유_2순위",
        "reason3_col": "A.[국내] Q3. 여행지 선택 이유_3순위",
        "transport_col": "국내]  B_여행1차_주요 이동 교통 수단(지역간)",
        "stay_price":      "A.[국내] Q8. 3. 숙박비 총액[단위:원]"
    }
}

def assert_columns_exist(df, cols, year):
    missing = [c for c in cols if c not in df.columns]
    if missing:
        raise KeyError(f"[{year}] 다음 컬럼이 데이터프레임에 없습니다: {missing}")

for year, conf in col_config.items():
    df = conf["df"].copy()

    # 필요한 모든 원본 컬럼 목록
    required_cols = [
        conf["region_col"], conf["age_col"], conf["sex_col"],
        conf["group_col"], conf["spend_col"], conf["stay_col"],
        conf["stay_day_col"], conf["total_cost_col"], conf["travel_type_col"],
        conf["reason1_col"], conf["reason2_col"], conf["reason3_col"],
        conf["transport_col"], conf["stay_price"]
    ]
    assert_columns_exist(df, required_cols, year)

    # 숫자형 변환 컬럼들(에러는 NaN으로) — 분석에 쓰는 항목 위주
    numeric_cols = [
        conf["group_col"], conf["age_col"], conf["spend_col"],
        conf["stay_day_col"], conf["total_cost_col"], conf["stay_price"]
    ]
    for c in numeric_cols:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    # 소규모 여행자(1~4인) 필터
    df_filtered = df[df[conf["group_col"]].between(1, 4)].copy()

    # 필요한 컬럼만 선택
    selected_cols = [
        conf["region_col"], conf["age_col"], conf["sex_col"],
        conf["group_col"], conf["spend_col"], conf["stay_col"],
        conf["stay_day_col"], conf["total_cost_col"], conf["stay_price"],
        conf["travel_type_col"], conf["reason1_col"], conf["reason2_col"],
        conf["reason3_col"], conf["transport_col"]
    ]
    df_final = df_filtered[selected_cols].copy()

    # 컬럼명 통일 (숙박비 총액 포함)
    df_final.columns = [
        "지역코드", "연령", "성별", "여행인원", "1인지출", "숙박형태", "여행일수",
        "총경비", "숙박비총액", "여행유형", "선택이유1", "선택이유2", "선택이유3", "교통수단"
    ]

    # 연도 컬럼 추가
    df_final["연도"] = year

    # CSV 저장
    out_path = f"소규모여행자_{year}.csv"
    df_final.to_csv(out_path, index=False, encoding="utf-8-sig")
    print(f"[{year}] 저장 완료: {out_path}, 행수={len(df_final):,}")
```

```python
import pandas as pd
data22 = pd.read_csv("소규모여행자_2022.csv")
data23 = pd.read_csv("소규모여행자_2023.csv")
data24 = pd.read_csv("소규모여행자_2024.csv")
```

```python
# 대상 컬럼들
int_cols = ["지역코드", "성별", "여행인원", "여행유형", "여행일수", "연령"]

# 전처리 함수
def convert_to_int(df):
    for col in int_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(-1).astype(int)
    return df

# 적용
data22 = convert_to_int(data22)
data23 = convert_to_int(data23)
data24 = convert_to_int(data24)
```

```python
def print_stay_stats(df, year):
    # 여행일수 숫자형으로 변환
    df["여행일수"] = pd.to_numeric(df["여행일수"], errors="coerce")

    total = len(df)
    daytrip = (df["여행일수"] == 0).sum()
    stay = (df["여행일수"] >= 1).sum()
    
    print(f"[{year}] 전체 소규모 여행자 수: {total:,}명")
    print(f" - 당일치기 여행자(0박): {daytrip:,}명 ({daytrip / total * 100:.1f}%)")
    print(f" - 숙박 여행자(1박 이상): {stay:,}명 ({stay / total * 100:.1f}%)\n")

print_stay_stats(data22, 2022)
print_stay_stats(data23, 2023)
print_stay_stats(data24, 2024)
```

```python
숙박_mean = df[df['숙박여부']=="숙박"]['1인지출'].mean()
당일_mean = df[df['숙박여부']=="당일치기"]['1인지출'].mean()
```

```python
# ===== 숙박비 제외한 지출 비교 =====
# 숙박여행객 / 당일치기 데이터 분리
overnight = df[df['숙박여부'] == "숙박"].copy()
daytrip = df[df['숙박여부'] == "당일치기"].copy()

# 1인 총 숙박비 계산: 숙박비총액 / 여행인원
overnight["lod_pp_total"] = overnight["숙박비총액"] / overnight["여행인원"]

# 비숙박 지출 계산
overnight_non_lodging_mean = (overnight["1인지출"] - overnight["lod_pp_total"]).mean()
daytrip_non_lodging_mean   = daytrip["1인지출"].mean()  # 당일은 숙박비가 없음

# 그래프
plt.bar(['숙박(숙박비 제외)', '당일치기'], 
        [overnight_non_lodging_mean, daytrip_non_lodging_mean], 
        color=['skyblue', 'salmon'])
plt.ylabel('평균 1인 지출 (원)')
plt.title('숙박비 제외 시 숙박 vs 당일치기 평균 1인 지출 비교')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 값 표시
for i, v in enumerate([overnight_non_lodging_mean, daytrip_non_lodging_mean]):
    plt.text(i, v + 2000, f"{v:,.0f}원", ha='center', fontweight='bold')

plt.show()
```

```python
# ===== 비숙박 지출 비교: 버전 A(전체 숙박비 제외) & 버전 B(1박 숙박비만 제외) =====
import numpy as np

# 1) 표본 분리
overnight = df[df['숙박여부'] == "숙박"].copy()
daytrip  = df[df['숙박여부'] == "당일치기"].copy()

# 2) 숫자형 안전 변환
for c in ['1인지출','여행일수','여행인원','숙박비총액']:
    df[c] = pd.to_numeric(df[c], errors='coerce')
    overnight[c] = pd.to_numeric(overnight[c], errors='coerce')
    daytrip[c] = pd.to_numeric(daytrip[c], errors='coerce')

# 3) 1인 숙박비 계산
#    - 전체 숙박비(여행기간 합) 1인 기준
overnight['lod_pp_total'] = np.where(overnight['여행인원'] >= 1,
                                     overnight['숙박비총액'] / overnight['여행인원'],
                                     np.nan)
#    - 1박 숙박비 1인 기준
overnight['lod_pp_per_night'] = np.where((overnight['여행인원'] >= 1) & (overnight['여행일수'] >= 1),
                                         (overnight['숙박비총액'] / overnight['여행인원']) / overnight['여행일수'],
                                         np.nan)

# 4) 비숙박 지출(평균) 계산
#    버전 A: "숙박비 전체"를 제외한 비숙박 지출 평균
overnight_nonlod_A = (overnight['1인지출'] - overnight['lod_pp_total']).where(lambda x: x >= 0)  # 음수는 NaN 처리
overnight_nonlod_A_mean = overnight_nonlod_A.mean()

#    버전 B: "1박 숙박비만" 제외한 비숙박 지출 평균
overnight_nonlod_B = (overnight['1인지출'] - overnight['lod_pp_per_night']).where(lambda x: x >= 0)
overnight_nonlod_B_mean = overnight_nonlod_B.mean()

#    당일치기는 숙박비가 0 → 두 버전 모두 동일(= 당일 1인지출 평균)
daytrip_nonlod_mean = daytrip['1인지출'].mean()

# 5) 시각화 (1행 2열)
fig, axes = plt.subplots(1, 2, figsize=(12, 4), constrained_layout=True)

#   (왼) 버전 A: 전체 숙박비 제외
axes[0].bar(['숙박(숙박비 전체 제외)', '당일치기'],
            [overnight_nonlod_A_mean, daytrip_nonlod_mean])
axes[0].set_title('비숙박 지출 비교 — 숙박비 전체 제외')
axes[0].set_ylabel('평균 1인 지출 (원)')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)
for i, v in enumerate([overnight_nonlod_A_mean, daytrip_nonlod_mean]):
    axes[0].text(i, v + 2000, f"{v:,.0f}원", ha='center', fontweight='bold')

#   (오) 버전 B: 1박 숙박비만 제외
axes[1].bar(['숙박(1박 숙박비 제외)', '당일치기'],
            [overnight_nonlod_B_mean, daytrip_nonlod_mean])
axes[1].set_title('비숙박 지출 비교 — 1박 숙박비만 제외')
axes[1].grid(axis='y', linestyle='--', alpha=0.7)
for i, v in enumerate([overnight_nonlod_B_mean, daytrip_nonlod_mean]):
    axes[1].text(i, v + 2000, f"{v:,.0f}원", ha='center', fontweight='bold')

plt.show()

# 6) 수치 요약도 함께 출력 (보고서용)
print("[요약] 비숙박 지출 평균")
print(f"- 숙박(숙박비 전체 제외): {overnight_nonlod_A_mean:,.0f} 원")
print(f"- 숙박(1박 숙박비만 제외): {overnight_nonlod_B_mean:,.0f} 원")
print(f"- 당일치기: {daytrip_nonlod_mean:,.0f} 원")
print(f"> 차이 A(숙박 전체제외 - 당일): {(overnight_nonlod_A_mean - daytrip_nonlod_mean):,.0f} 원")
print(f"> 차이 B(숙박 1박제외 - 당일): {(overnight_nonlod_B_mean - daytrip_nonlod_mean):,.0f} 원")
```

```python
import matplotlib.pyplot as plt

# 선택 이유 매핑
reason_map = {
    1: "지명도", 2: "볼거리", 3: "저렴한 경비", 4: "이동거리", 5: "여행 시간",
    6: "숙박시설", 7: "쇼핑", 8: "음식", 9: "교통편", 10: "체험 프로그램",
    11: "추천", 12: "편의시설", 13: "교육성", 14: "동반자", 15: "기타"
}

# 교통수단 매핑
transport_map = {
    1: "자가용", 2: "철도", 3: "항공기", 4: "선박", 5: "지하철", 6: "정기버스",
    7: "전세버스", 8: "렌트", 9: "자전거", 10: "택시", 11: "기타"
}

# 데이터 리스트와 연도 리스트
data_list = [data22, data23, data24]
year_list = [2022, 2023, 2024]

# 시각화 함수
def process_and_plot(df, year):
    # 선택이유 매핑
    for i in range(1, 4):
        col = f"선택이유{i}"
        df[col] = pd.to_numeric(df[col], errors="coerce").map(reason_map)

    # 교통수단 매핑
    df["교통수단"] = pd.to_numeric(df["교통수단"], errors="coerce").map(transport_map)

    # 시각화
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    for i in range(1, 4):
        col = f"선택이유{i}"
        counts = df[col].value_counts().sort_values(ascending=False)
        ax = axes[(i-1)//2, (i-1)%2]
        counts.plot(kind="bar", ax=ax)
        ax.set_title(f"{year}년 선택 이유 {i}순위")
        ax.set_ylabel("응답 수")
        ax.set_xlabel("이유")
        ax.tick_params(axis='x', rotation=45)

    # 교통수단
    counts = df["교통수단"].value_counts().sort_values(ascending=False)
    ax = axes[1, 1]
    counts.plot(kind="bar", ax=ax, color="skyblue")
    ax.set_title(f"{year}년 주요 교통수단")
    ax.set_ylabel("응답 수")
    ax.set_xlabel("교통수단")
    ax.tick_params(axis='x', rotation=45)

    plt.suptitle(f"{year}년 선택 이유 및 교통수단 분포", fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# 실행
for df, year in zip(data_list, year_list):
    process_and_plot(df.copy(), year)
```

```python
# ===== 통합 파이프라인 (정리版): K 비교 → k=3 학습 → 분포/특성/산점도 → 자동 라벨링 =====
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 0) 필수 컬럼 체크(안전장치)
REQUIRED_COLS = ["1인지출","여행일수","여행인원","숙박여부","숙박형태","교통수단","여행유형_그룹"]
def assert_required_cols(df, name="df"):
    missing = [c for c in REQUIRED_COLS if c not in df.columns]
    if missing:
        raise KeyError(f"[{name}] 누락 컬럼: {missing}")

# 1) 전처리 → X_scaled 만들기
def build_X_scaled(df):
    df = df.copy()
    assert_required_cols(df, "입력데이터")

    # 숙박여부 이진화
    df["숙박여부"] = df["숙박여부"].map({"숙박": 1, "당일치기": 0})
    df = df[pd.to_numeric(df["숙박여부"], errors='coerce').notnull()]
    df["숙박여부"] = df["숙박여부"].astype(float)

    # 사용 변수
    numeric_cols = ["1인지출", "여행일수", "여행인원", "숙박여부"]
    categorical_cols = ["숙박형태", "교통수단", "여행유형_그룹"]

    # 원핫 인코딩(버전 호환)
    try:
        enc = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
    except TypeError:
        enc = OneHotEncoder(sparse=False, handle_unknown="ignore")

    enc_arr = enc.fit_transform(df[categorical_cols])
    enc_df = pd.DataFrame(enc_arr, columns=enc.get_feature_names_out(categorical_cols), index=df.index)

    # 결합 + 스케일링
    X = pd.concat([df[numeric_cols], enc_df], axis=1)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    return df, X_scaled, enc, scaler, X.columns.tolist()

# 2) k 값 비교(실루엣)
def silhouette_by_k(X_scaled, k_range=range(2,7), random_state=42):
    scores = {}
    for k in k_range:
        labels = KMeans(n_clusters=k, n_init=10, random_state=random_state).fit_predict(X_scaled)
        scores[k] = silhouette_score(X_scaled, labels)
    return scores

# 3) KMeans 학습
def kmeans_with_labels(df, X_scaled, k=3, random_state=42):
    km = KMeans(n_clusters=k, n_init=10, random_state=random_state)
    labels = km.fit_predict(X_scaled)
    df_out = df.copy()
    df_out["클러스터"] = labels
    return df_out, km

# 4) 시각화: 분포/특성/PCA 산점도
def plot_cluster_counts(df, year):
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(data=df, x="클러스터", hue="클러스터", palette="Set2", legend=False)
    for p in ax.patches:
        h = int(p.get_height())
        ax.annotate(f"{h}", (p.get_x()+p.get_width()/2, h), ha="center", va="bottom", fontweight="bold")
    plt.title(f"{year}년 클러스터별 인원 분포")
    plt.xlabel("클러스터"); plt.ylabel("인원 수")
    plt.tight_layout(); plt.show()

def plot_cluster_features(df, year):
    cluster_summary = df.groupby("클러스터").agg({
        "1인지출": "mean", "여행일수": "mean", "여행인원": "mean", "숙박여부": "mean"
    }).reset_index()
    features = ["1인지출", "여행일수", "여행인원", "숙박여부"]
    fig, axs = plt.subplots(2, 2, figsize=(12, 8)); axs = axs.flatten()
    for i, feat in enumerate(features):
        sns.barplot(data=cluster_summary, x="클러스터", y=feat, hue="클러스터",
                    palette="Set2", ax=axs[i], legend=False)
        axs[i].set_title(f"{year}년 클러스터별 평균 {feat}")
        axs[i].set_xlabel("클러스터"); axs[i].set_ylabel("평균")
    plt.tight_layout(); plt.show()

def plot_cluster_scatter(X_scaled, labels, km_model, year):
    pca = PCA(n_components=2, random_state=42)
    X2 = pca.fit_transform(X_scaled)
    centers2 = pca.transform(km_model.cluster_centers_)

    plt.figure(figsize=(7,6))
    scatter = plt.scatter(X2[:,0], X2[:,1], c=labels, cmap="tab10", alpha=0.6, s=15)
    plt.scatter(centers2[:,0], centers2[:,1], c=range(km_model.n_clusters), cmap="tab10",
                marker="X", s=200, edgecolor="black", linewidths=1.2, label="centroid")
    plt.title(f"{year}년 클러스터 전체 산점도 (PCA 2D)")
    plt.xlabel("PC1"); plt.ylabel("PC2")
    plt.legend(*scatter.legend_elements(), title="클러스터", loc="best")
    plt.grid(alpha=0.3, linestyle="--")
    plt.tight_layout(); plt.show()

# 5) 자동 라벨링(해석 가능한 이름 부여)
def auto_label_clusters(df_labeled):
    """
    규칙:
      - 숙박여부 평균이 낮은(0.5 미만) 군집 중 1인지출이 가장 낮은 군집 → '저비용 당일치기형'
      - 숙박여부 평균이 높은 군집 중 1인지출이 가장 높은 군집 → '고지출 숙박형'
      - 나머지 → '균형형'
    """
    g = df_labeled.groupby("클러스터")[["숙박여부","1인지출","여행일수"]].mean()

    day_candidates = g[g["숙박여부"] < 0.5]
    day_label = day_candidates["1인지출"].idxmin() if len(day_candidates) else g["1인지출"].idxmin()

    stay_candidates = g[g["숙박여부"] >= 0.5]
    high_spend_stay = stay_candidates["1인지출"].idxmax() if len(stay_candidates) else g["1인지출"].idxmax()

    labels_map = {}
    for c in g.index:
        if c == day_label:
            labels_map[c] = "저비용 당일치기형"
        elif c == high_spend_stay:
            labels_map[c] = "고지출 숙박형"
        else:
            labels_map[c] = "균형형"

    out = df_labeled.copy()
    out["클러스터유형"] = out["클러스터"].map(labels_map)
    return out, labels_map

# 6) 실행 함수(연도별 일괄)
def run_all(year, data, k=3, k_range=range(2,7)):
    # 전처리
    df_proc, X_scaled, enc, scaler, feat_names = build_X_scaled(data)

    # k 비교
    scores = silhouette_by_k(X_scaled, k_range)
    print(f"{year}년 k 값별 실루엣:", scores)

    # k 적용
    df_labeled, km = kmeans_with_labels(df_proc, X_scaled, k=k)

    # 자동 라벨링
    df_named, mapping = auto_label_clusters(df_labeled)
    print(f"{year}년 자동 라벨 매핑:", mapping)

    # 시각화
    plot_cluster_counts(df_named, year)
    plot_cluster_features(df_named, year)
    plot_cluster_scatter(X_scaled, df_named["클러스터"].values, km, year)

    return df_named, {"scores": scores, "mapping": mapping}

# ===== 실행 =====
# 전제: data22, data23, data24가 메모리에 있고 필수 컬럼을 포함해야 함
data22_clustered, info22 = run_all(2022, data22, k=3, k_range=range(2,7))
data23_clustered, info23 = run_all(2023, data23, k=3, k_range=range(2,7))
data24_clustered, info24 = run_all(2024, data24, k=3, k_range=range(2,7))

# 결과 확인 예시
print(data22_clustered[["클러스터","클러스터유형"]].head())
print(data23_clustered[["클러스터","클러스터유형"]].head())
print(data24_clustered[["클러스터","클러스터유형"]].head())
```

```python
# =========================================
# 전체 수정 코드 (안전 전처리 + k=3 클러스터링 + 4분할 시각화)
# =========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# --------------------------------------------------
# 0) 필수 컬럼 체크
# --------------------------------------------------
REQUIRED_COLS = ["1인지출","여행일수","여행인원","숙박여부","숙박형태","교통수단","여행유형_그룹"]

def assert_required_cols(df, name="df"):
    missing = [c for c in REQUIRED_COLS if c not in df.columns]
    if missing:
        raise KeyError(f"[{name}] 누락 컬럼: {missing}")

# --------------------------------------------------
# 1) 안전 전처리 → X_scaled 만들기 (★수정: 범주형을 문자열화 + NaN→'미상')
# --------------------------------------------------
def build_X_scaled(df):
    df = df.copy()
    assert_required_cols(df, "입력데이터")

    # 숫자형 안전 변환
    # 숙박여부: 문자열/숫자 혼재 → 1/0로 통일
    if df["숙박여부"].dtype == object:
        df["숙박여부"] = df["숙박여부"].map({"숙박":1, "당일치기":0})
    df["숙박여부"] = pd.to_numeric(df["숙박여부"], errors="coerce").fillna(0).clip(0,1).astype(float)

    for c in ["1인지출","여행일수","여행인원"]:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    # 범주형 안전 변환: 전부 문자열화 + 결측·이상값 → '미상'
    categorical_cols = ["숙박형태","교통수단","여행유형_그룹"]
    for c in categorical_cols:
        df[c] = df[c].astype(str)  # NaN -> 'nan'
        df[c] = df[c].replace({"nan":"미상", "None":"미상", "NaN":"미상"}).str.strip()
        df[c] = df[c].replace({"": "미상", " ": "미상"})

    numeric_cols = ["1인지출","여행일수","여행인원","숙박여부"]

    # 원-핫 인코딩 (버전 호환)
    try:
        enc = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
    except TypeError:
        enc = OneHotEncoder(sparse=False, handle_unknown="ignore")
    enc_arr = enc.fit_transform(df[categorical_cols])
    enc_df = pd.DataFrame(enc_arr, columns=enc.get_feature_names_out(categorical_cols), index=df.index)

    # 결합 + 스케일링
    X = pd.concat([df[numeric_cols], enc_df], axis=1).fillna(0)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # 후속 매칭을 위해 인덱스 리셋
    df.reset_index(drop=True, inplace=True)

    return df, X_scaled, enc, scaler, X.columns.tolist()

# --------------------------------------------------
# 2) k 값 비교(옵션)
# --------------------------------------------------
def silhouette_by_k(X_scaled, k_range=range(2,7), random_state=42):
    scores = {}
    for k in k_range:
        labels = KMeans(n_clusters=k, n_init=10, random_state=random_state).fit_predict(X_scaled)
        scores[k] = silhouette_score(X_scaled, labels)
    return scores

# --------------------------------------------------
# 3) KMeans 학습 + 라벨 부여
# --------------------------------------------------
def kmeans_with_labels(df, X_scaled, k=3, random_state=42):
    km = KMeans(n_clusters=k, n_init=10, random_state=random_state)
    labels = km.fit_predict(X_scaled)
    df_out = df.copy()
    df_out["클러스터"] = labels
    return df_out, km

# --------------------------------------------------
# 4) 자동 라벨링(해석 가능한 이름)
#   - 숙박여부 평균이 낮으면서 지출도 낮은 군집 → '저비용 당일치기형'
#   - 숙박여부 평균이 높은 군집 중 지출이 가장 높은 군집 → '고지출 숙박형'
#   - 나머지 → '균형형'
# --------------------------------------------------
def auto_label_clusters(df_labeled):
    g = df_labeled.groupby("클러스터")[["숙박여부","1인지출","여행일수"]].mean()

    day_candidates = g[g["숙박여부"] < 0.5]
    day_label = day_candidates["1인지출"].idxmin() if len(day_candidates) else g["1인지출"].idxmin()

    stay_candidates = g[g["숙박여부"] >= 0.5]
    high_spend_stay = stay_candidates["1인지출"].idxmax() if len(stay_candidates) else g["1인지출"].idxmax()

    labels_map = {}
    for c in g.index:
        if c == day_label:
            labels_map[c] = "저비용 당일치기형"
        elif c == high_spend_stay:
            labels_map[c] = "고지출 숙박형"
        else:
            labels_map[c] = "균형형"

    out = df_labeled.copy()
    out["클러스터유형"] = out["클러스터"].map(labels_map)
    return out, labels_map

# --------------------------------------------------
# 5) 연도별 실행 함수 (k 선택 → 학습 → 자동 라벨)
# --------------------------------------------------
def run_all(year, data, k=3, k_range=range(2,7)):
    df_proc, X_scaled, enc, scaler, feat_names = build_X_scaled(data)
    # (옵션) 실루엣 확인
    # scores = silhouette_by_k(X_scaled, k_range)
    # print(f"{year}년 k 값별 실루엣:", scores)

    df_labeled, km = kmeans_with_labels(df_proc, X_scaled, k=k)
    df_named, mapping = auto_label_clusters(df_labeled)
    # print(f"{year}년 자동 라벨 매핑:", mapping)
    return df_named, {"mapping": mapping}

# --------------------------------------------------
# 6) 4분할 레이더 차트
# --------------------------------------------------
CLUSTER_TYPE_COLORS = {
    "고지출 숙박형": "#2E86DE",
    "저비용 당일치기형": "#EB984E",
    "균형형": "#7DCEA0"
}
FEATURES = ("1인지출","여행일수","여행인원","숙박여부")

def plot_radar_grid_2x2(named_by_year: dict, title="연도별 군집 특성 비교 (정규화)"):
    fig, axes = plt.subplots(2, 2, subplot_kw={"polar": True}, figsize=(12, 9))
    axes = axes.flatten()

    labels = list(FEATURES)
    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
    angles = np.concatenate([angles, angles[:1]])

    for ax, (y, df_named) in zip(axes, named_by_year.items()):
        summary = df_named.groupby(["클러스터","클러스터유형"])[list(FEATURES)].mean().reset_index()

        # 패널 내부 정규화 (모양 비교 목적)
        norm = summary.copy()
        for f in FEATURES:
            mn, mx = summary[f].min(), summary[f].max()
            norm[f] = 0.5 if mx == mn else (summary[f] - mn) / (mx - mn)

        for _, row in norm.iterrows():
            vals = row[list(FEATURES)].values
            vals = np.concatenate([vals, vals[:1]])
            ctype = row["클러스터유형"]
            color = CLUSTER_TYPE_COLORS.get(ctype, None)
            label = f'{int(row["클러스터"])} ({ctype})'
            ax.plot(angles, vals, linewidth=2, color=color, label=label)
            ax.fill(angles, vals, alpha=0.10, color=color)

        ax.set_xticks(angles[:-1])
        ax.set_xticklabels(labels, fontsize=10)
        ax.set_yticklabels([])
        ax.set_title(str(y), pad=10, fontsize=12)
        ax.grid(alpha=0.25, linestyle="--")

    # 하나만 범례 표시(오른쪽 위)
    axes[1].legend(loc="upper right", bbox_to_anchor=(1.35, 1.10), title="클러스터(유형)", fontsize=9)
    fig.suptitle(title, fontsize=14, y=0.98)
    plt.tight_layout()
    plt.show()

# --------------------------------------------------
# 7) 4분할 PCA 산점도 (★수정: 인덱스 정렬 일치 + 안전 전처리)
# --------------------------------------------------
def plot_pca_grid_2x2(raw_by_year: dict, named_by_year: dict, k=3, title="연도별 클러스터 분포 (PCA 2D)"):
    fig, axes = plt.subplots(2, 2, figsize=(12, 9))
    axes = axes.flatten()

    # 고정 팔레트
    base_colors = {0:"#2E86DE", 1:"#EB984E", 2:"#7DCEA0", 3:"#AF7AC5", 4:"#48C9B0"}

    for ax, y in zip(axes, raw_by_year.keys()):
        raw_df   = raw_by_year[y]
        df_named = named_by_year[y]

        # 동일 전처리
        df_proc, X_scaled, _, _, _ = build_X_scaled(raw_df)

        # 순서 일치 보장
        df_proc  = df_proc.reset_index(drop=True)
        df_named = df_named.reset_index(drop=True)

        labels = df_named["클러스터"].values
        # types  = df_named["클러스터유형"].values  # 필요시 사용

        # PCA 투영
        pca = PCA(n_components=2, random_state=42)
        X2 = pca.fit_transform(X_scaled)

        # 중심점(옵션: 동일 k-means 재학습)
        km = KMeans(n_clusters=k, n_init=10, random_state=42).fit(X_scaled)
        centers2 = pca.transform(km.cluster_centers_)

        point_colors = [base_colors.get(int(c), "#7f8c8d") for c in labels]
        ax.scatter(X2[:,0], X2[:,1], c=point_colors, s=14, alpha=0.65, edgecolors="none")
        ax.scatter(centers2[:,0], centers2[:,1], c=[base_colors.get(i,"#7f8c8d") for i in range(k)],
                   marker="X", s=160, edgecolors="black", linewidths=1.0)

        # 간단 범례: id→유형
        id_to_type = df_named.groupby("클러스터")["클러스터유형"].first().to_dict()
        handles = [
            plt.Line2D([0],[0], marker='o', color='w',
                       label=f"{cid} ({t})",
                       markerfacecolor=base_colors.get(cid,"#7f8c8d"), markersize=8)
            for cid, t in id_to_type.items()
        ]
        handles.append(plt.Line2D([0],[0], marker='X', color='w',
                                  label='centroid', markerfacecolor='gray',
                                  markeredgecolor='black', markersize=8))
        ax.legend(handles=handles, loc="best", fontsize=8)

        ax.set_title(str(y), fontsize=12)
        ax.set_xlabel("PC1"); ax.set_ylabel("PC2")
        ax.grid(alpha=0.3, linestyle="--")

    fig.suptitle(title, fontsize=14, y=0.98)
    plt.tight_layout()
    plt.show()

# ==================================================
# 사용 예시
#  - 전제: data22, data23, data24 가 이미 메모리에 있고 REQUIRED_COLS 포함
# ==================================================

# (A) 연도별 클러스터링 실행 (k=3 고정)
data22_clustered, info22 = run_all(2022, data22, k=3)
data23_clustered, info23 = run_all(2023, data23, k=3)
data24_clustered, info24 = run_all(2024, data24, k=3)

# (B) '전체' 데이터/라벨 만들기
data_all = pd.concat([data22, data23, data24], ignore_index=True)
named_all = pd.concat([
    data22_clustered.assign(__year="2022"),
    data23_clustered.assign(__year="2023"),
    data24_clustered.assign(__year="2024"),
], ignore_index=True)

# (C) dict 구성(4분할용)
named_by_year = {
    "2022": data22_clustered,
    "2023": data23_clustered,
    "2024": data24_clustered,
    "전체":  named_all
}
raw_by_year = {
    "2022": data22,
    "2023": data23,
    "2024": data24,
    "전체":  data_all
}

# (D) 4분할 시각화 실행
plot_radar_grid_2x2(named_by_year, title="연도별 군집 특성 비교 (정규화)")
plot_pca_grid_2x2(raw_by_year, named_by_year, k=3, title="연도별 클러스터 분포 (PCA 2D)")
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples

def silhouette_by_k(X_scaled, k_range=range(2,7), random_state=42):
    """k 값을 바꿔가며 전체 실루엣 점수를 계산"""
    scores = {}
    for k in k_range:
        labels = KMeans(n_clusters=k, n_init=10, random_state=random_state).fit_predict(X_scaled)
        scores[k] = silhouette_score(X_scaled, labels)
    return scores

def silhouette_summary(X_scaled, labels):
    """전체 평균, 군집별 평균/개수 요약"""
    s_all = silhouette_score(X_scaled, labels)
    s_samples = silhouette_samples(X_scaled, labels)
    df = pd.DataFrame({"label": labels, "s": s_samples})
    by = df.groupby("label")["s"].agg(["mean","count"]).reset_index()
    return s_all, by

def plot_silhouette_scores(scores, title="Silhouette vs k"):
    """k별 실루엣 점수 라인차트"""
    ks = sorted(scores.keys())
    vals = [scores[k] for k in ks]
    plt.figure(figsize=(5,3.2))
    plt.plot(ks, vals, marker="o")
    plt.xticks(ks)
    plt.ylabel("Silhouette score")
    plt.xlabel("k")
    plt.title(title)
    plt.grid(alpha=0.3, linestyle="--")
    plt.tight_layout()
    plt.show()
```

```python
def run_all(year, data, k=3, k_range=range(2,7)):
    df_proc, X_scaled, enc, scaler, feat_names = build_X_scaled(data)

    # k별 점수(옵션) & k=3 점수
    scores = silhouette_by_k(X_scaled, k_range=k_range)
    df_labeled, km = kmeans_with_labels(df_proc, X_scaled, k=k)
    df_named, mapping = auto_label_clusters(df_labeled)

    # k=3 점수 요약
    s_overall, s_by = silhouette_summary(X_scaled, df_named["클러스터"].values)
    info = {
        "mapping": mapping,
        "silhouette_overall": s_overall,
        "silhouette_by_cluster": s_by,
        "silhouette_k_scores": scores,   # {k: score}
    }
    return df_named, info
```

```python
# (A) 연도별 클러스터링 실행 (k=3 고정)
data22_clustered, info22 = run_all(2022, data22, k=3)
data23_clustered, info23 = run_all(2023, data23, k=3)
data24_clustered, info24 = run_all(2024, data24, k=3)

# (A-1) 연도별 실루엣 점수 간단 출력
for y, info in [("2022", info22), ("2023", info23), ("2024", info24)]:
    print(f"=== {y}년 실루엣(전체) = {info['silhouette_overall']:.3f}")
    print(info["silhouette_by_cluster"], "\n")  # 군집별 평균/개수

# (A-2) k=2~6 탐색 그래프(연도별 1장씩 필요 시)
plot_silhouette_scores(info22["silhouette_k_scores"], title="2022 Silhouette vs k")
plot_silhouette_scores(info23["silhouette_k_scores"], title="2023 Silhouette vs k")
plot_silhouette_scores(info24["silhouette_k_scores"], title="2024 Silhouette vs k")
```

```python
grouped_dfs = []
for year, df in zip([2022, 2023, 2024], [data22_clustered, data23_clustered, data24_clustered]):
    tmp = (
        df[df["클러스터유형"] == "고지출 숙박형"]
        .groupby("시도명")
        .size()
        .reset_index(name="고지출숙박형_수")
    )
    tmp["연도"] = year
    grouped_dfs.append(tmp)

cluster_demand = pd.concat(grouped_dfs, ignore_index=True)
cluster_demand
```

```python
# 시도별 요약 지표 계산
summary_dfs = []
for df in [data22_clustered, data23_clustered, data24_clustered]:
    temp = (
        df.groupby(["연도", "시도명"])
        .agg(
            숙박률=("숙박여부", lambda x: (x == 1).mean()),
            평균지출비용=("1인지출", "mean"),
            여행자수=("지역코드", "count"),
            대표_지역코드=("지역코드", lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
        )
        .reset_index()
    )
    summary_dfs.append(temp)

summary_by_region = pd.concat(summary_dfs, ignore_index=True)

# 수요 데이터와 결합하여 infra_base 생성
infra_base = pd.merge(
    cluster_demand,
    summary_by_region,
    on=["연도", "시도명"],
    how="left"
)
```

```python
# 기준값 계산
threshold_demand = infra_base["고지출숙박형_수"].quantile(0.5)
threshold_stay = infra_base["숙박률"].median()

# 조건 추가
infra_base["수요있음"] = infra_base["고지출숙박형_수"] >= threshold_demand
infra_base["저활용여부"] = infra_base["숙박률"] < threshold_stay

# 입지 후보 필터링
infra_candidates = infra_base[
    (infra_base["수요있음"]) & (infra_base["저활용여부"])
].copy()
```

```python
import pandas as pd

# 지역코드 ↔ 지역명 매핑
region_map = {
    31370: "가평", 36020: "여수", 37020: "경주",
    35010: "전주", 33380: "단양", 34030: "보령"
}
reverse_map = {v: k for k, v in region_map.items()}

# 원본 복사 및 지역명 추가
df = top3_by_region_year.copy()
df["지역명"] = df["지역코드"].map(region_map)

# 가족/친지집 제외 후 Top3 추출 함수 정의
def get_top3_excluding(group):
    제외형태 = ["가족/친지집"]
    df_filtered = group[~group["숙박형태명"].isin(제외형태)].copy()
    df_filtered = df_filtered.sort_values("빈도", ascending=False).head(3)
    df_filtered["순위"] = range(1, len(df_filtered) + 1)
    return df_filtered

# 연도-지역명 단위로 Top3 추출
group_cols = ["연도", "지역명"]
filtered_df = (
    df.set_index(group_cols)
      .groupby(level=group_cols, group_keys=False)
      .apply(get_top3_excluding)
      .reset_index()
)

# 피벗: 숙박형태명 + 빈도 분리
pivot_df = filtered_df.pivot(index=["연도", "지역명"], columns="순위", values=["숙박형태명", "빈도"])
pivot_df.columns = [f"Top{i}_{col}" for col, i in pivot_df.columns]
pivot_df = pivot_df.reset_index()

# 지역코드 복원
pivot_df["지역코드"] = pivot_df["지역명"].map(reverse_map)
```

```python
# 여행자수 및 평균 숙박일수 계산
region_summary = []
for _, row in pivot_df.iterrows():
    year = row["연도"]
    code = row["지역코드"]
    subset = full_data[
        (full_data["연도"] == year) &
        (full_data["지역코드"] == code)
    ]
    region_summary.append({
        "연도": year,
        "지역코드": code,
        "여행자수": len(subset),
        "평균숙박일수": round(subset["여행일수"].mean(), 2) if not subset.empty else 0
    })

region_summary_df = pd.DataFrame(region_summary)
region_summary_df
```

```python
import pandas as pd
import numpy as np

# 0) 안전 전처리: 숙박여부를 1/0으로 통일
def ensure_stay_flag(df):
    d = df.copy()
    if d["숙박여부"].dtype == object:
        mapping = {"숙박": 1, "당일치기": 0, "무응답": np.nan}
        d["숙박여부"] = d["숙박여부"].map(mapping)
    d["숙박여부"] = pd.to_numeric(d["숙박여부"], errors="coerce")
    return d

# 1) 세 연도 합치기 (필요 컬럼만)
base22 = ensure_stay_flag(data22[["연도","시도명","지역코드","숙박여부"]])
base23 = ensure_stay_flag(data23[["연도","시도명","지역코드","숙박여부"]])
base24 = ensure_stay_flag(data24[["연도","시도명","지역코드","숙박여부"]])

all_base = pd.concat([base22, base23, base24], ignore_index=True)

# 2) 여수(36020)만 추출 후 숙박률 계산
def compute_stay_rate(df, region_code=36020):
    d = df[df["지역코드"] == region_code].copy()
    if d.empty:
        return pd.DataFrame(columns=["연도","전체여행자수","숙박자수","비숙박자수","숙박률(%)"])
    grp = (
        d.groupby("연도")["숙박여부"]
         .agg(숙박자수="sum", 전체여행자수="count")
         .reset_index()
    )
    grp["비숙박자수"] = grp["전체여행자수"] - grp["숙박자수"]
    grp["숙박률(%)"] = (grp["숙박자수"] / grp["전체여행자수"] * 100).round(1)

    # 전체 합계 행 추가
    total = grp[["숙박자수","전체여행자수","비숙박자수"]].sum()
    total_rate = (total["숙박자수"] / total["전체여행자수"] * 100) if total["전체여행자수"]>0 else np.nan
    total_row = pd.DataFrame([{
        "연도": "전체",
        "전체여행자수": int(total["전체여행자수"]),
        "숙박자수": int(total["숙박자수"]),
        "비숙박자수": int(total["비숙박자수"]),
        "숙박률(%)": round(total_rate, 1) if pd.notna(total_rate) else np.nan
    }])
    return pd.concat([grp, total_row], ignore_index=True)

yeosu_stay_rate = compute_stay_rate(all_base, region_code=36020)
yeosu_stay_rate
```

```python
# 3) 고지출 숙박형 지표 계산 함수 (지출 포함)
def add_high_spend_cluster_info(
    yeosu_stay_rate,
    dfs_by_year,
    region_code=36020,
    high_label="고지출 숙박형"
):
    result = yeosu_stay_rate.copy()

    hs_list = []
    for year, df in dfs_by_year:
        hs_df = df[(df["지역코드"] == region_code) & (df["클러스터유형"] == high_label)].copy()
        hs_df["여행일수"] = pd.to_numeric(hs_df.get("여행일수"), errors="coerce")
        hs_df["1인지출"]  = pd.to_numeric(hs_df.get("1인지출"),  errors="coerce")

        if not hs_df.empty:
            # 연도별 총합(총박수/총지출)로 1박당지출 가중계산
            total_nights = hs_df["여행일수"].sum(skipna=True)
            total_spend  = hs_df["1인지출"].sum(skipna=True)
            pernight     = (total_spend / total_nights) if (pd.notna(total_nights) and total_nights > 0) else np.nan

            hs_list.append({
                "연도": year,
                "고지출숙박형_수": int(hs_df.shape[0]),
                "고지출_평균숙박일수": round(hs_df["여행일수"].mean(), 2),
                "고지출_평균지출비용": round(hs_df["1인지출"].mean(), 2),
                "고지출_1박당지출": round(pernight, 2)
            })
        else:
            hs_list.append({
                "연도": year,
                "고지출숙박형_수": 0,
                "고지출_평균숙박일수": np.nan,
                "고지출_평균지출비용": np.nan,
                "고지출_1박당지출": np.nan
            })

    # 전체 행(가중 평균)
    all_hs_df = pd.concat(
        [df[(df["지역코드"] == region_code) & (df["클러스터유형"] == high_label)] for _, df in dfs_by_year],
        ignore_index=True
    )
    all_hs_df["여행일수"] = pd.to_numeric(all_hs_df.get("여행일수"), errors="coerce")
    all_hs_df["1인지출"]  = pd.to_numeric(all_hs_df.get("1인지출"),  errors="coerce")

    if not all_hs_df.empty:
        total_nights = all_hs_df["여행일수"].sum(skipna=True)
        total_spend  = all_hs_df["1인지출"].sum(skipna=True)
        pernight     = (total_spend / total_nights) if (pd.notna(total_nights) and total_nights > 0) else np.nan

        hs_list.append({
            "연도": "전체",
            "고지출숙박형_수": int(all_hs_df.shape[0]),
            "고지출_평균숙박일수": round(all_hs_df["여행일수"].mean(), 2),
            "고지출_평균지출비용": round(all_hs_df["1인지출"].mean(), 2),
            "고지출_1박당지출": round(pernight, 2)
        })
    else:
        hs_list.append({
            "연도": "전체",
            "고지출숙박형_수": 0,
            "고지출_평균숙박일수": np.nan,
            "고지출_평균지출비용": np.nan,
            "고지출_1박당지출": np.nan
        })

    hs_df_final = pd.DataFrame(hs_list)
    return result.merge(hs_df_final, on="연도", how="left")


# 4) 실행
yeosu_stay_rate = add_high_spend_cluster_info(
    yeosu_stay_rate,
    dfs_by_year=[(2022, data22_clustered), (2023, data23_clustered), (2024, data24_clustered)],
    region_code=36020,
    high_label="고지출 숙박형"
)

yeosu_stay_rate
```

```python
import pandas as pd
import numpy as np

# ===== 입력 요약(이미 보유): yeosu_stay_rate에 아래 컬럼이 있어야 함 =====
# ["연도","고지출숙박형_수","고지출_평균숙박일수","고지출_1박당지출"]

# ===== 시나리오 정의 =====
# (시나리오명, ΔN(추가숙박일수/인), 인원증가율, 1박당지출 변화율, 주변 방문율, 주변 1인 지출)
scenarios = [
    ("보수", 1.0, 0.00, -0.05, 0.20, 60000),  # 방문율 20%, 6만원/인
    ("기준", 1.5, 0.05,  0.00, 0.30, 70000),  # 방문율 30%, 7만원/인
    ("확장", 2.0, 0.10,  0.05, 0.40, 80000),  # 방문율 40%, 8만원/인
]

def run_scenarios_with_indirect(df, scenarios):
    rows = []
    for name, dN, d_vis, d_spend, spill_rate, spill_spend in scenarios:
        for _, r in df.iterrows():
            year   = r["연도"]
            n0     = float(r["고지출숙박형_수"])
            perN0  = float(r["고지출_1박당지출"])
            baseN  = float(r["고지출_평균숙박일수"])

            if any(pd.isna(x) for x in [n0, perN0, baseN]): 
                continue

            # 시나리오 적용(인원/단가 조정)
            n_new    = n0 * (1 + d_vis)
            perN_new = perN0 * (1 + d_spend)

            # 1) 직접효과(여수 안): ΔN × 인원 × 1박당지출
            direct = dN * n_new * perN_new

            # 2) 간접효과(주변 시군): ΔN × 인원 × (주변 방문율) × (방문 1인 지출)
            #   - 해석: 체류일수가 늘어난 만큼 ‘하루 혹은 반나절’ 주변 시군 당일 방문이 유발된다고 가정
            indirect = dN * n_new * spill_rate * spill_spend

            rows.append({
                "시나리오": name, "연도": year,
                "기준평균숙박일수": round(baseN,2),
                "추가숙박일수(ΔN)": dN,
                "시나리오_인원": round(n_new),
                "시나리오_1박당지출(원)": round(perN_new),
                "주변방문율": spill_rate, "주변1인지출(원)": spill_spend,
                "직접효과(원)": round(direct),
                "간접효과(원)": round(indirect),
                "총효과(원)": round(direct + indirect)
            })
    out = pd.DataFrame(rows)

    # 합계(전체 행 중심으로 보기 용이하게)
    out["직접효과(억원)"] = (out["직접효과(원)"] / 1e8).round(2)
    out["간접효과(억원)"] = (out["간접효과(원)"] / 1e8).round(2)
    out["총효과(억원)"]   = (out["총효과(원)"]   / 1e8).round(2)

    # 연도별 요약 & '전체'만 발췌
    by_year   = out.groupby(["시나리오","연도"], as_index=False)[["직접효과(원)","간접효과(원)","총효과(원)"]].sum()
    by_year[["직접효과(억원)","간접효과(억원)","총효과(억원)"]] = (by_year[["직접효과(원)","간접효과(원)","총효과(원)"]] / 1e8).round(2)
    overall  = by_year[by_year["연도"]=="전체"].sort_values("시나리오")
    return out, by_year, overall[["시나리오","직접효과(억원)","간접효과(억원)","총효과(억원)"]]

# 실행
detail, yearly, overall = run_scenarios_with_indirect(yeosu_stay_rate, scenarios)

# 확인용 상위 표
overall  # 시나리오별 '전체' 기준: 여수 직접 + 주변 간접 + 총효과(억원)
```

```python
# 공영주차장 수 (자가용)

file_path = '전국주차장정보표준데이터.csv'
parking = pd.read_csv(file_path, encoding="euc-kr", low_memory=False)
selected = parking[["주차장명", "주차장구분", "소재지지번주소"]]
public_parking = selected[selected["주차장구분"] == "공영"]
```

```python
parking["주소"] = parking["소재지도로명주소"].fillna("")  
parking.loc[parking["주소"] == "", "주소"] = parking["소재지지번주소"]

def extract_sigungoo(address):
    try:
        parts = address.split()
        if len(parts) >= 2:
            return parts[0] + " " + parts[1]  
    except:
        return None

public = parking[parking["주차장구분"] == "공영"].copy()

public["시군구"] = public["주소"].apply(extract_sigungoo)

parking_counts = public["시군구"].value_counts().reset_index()
parking_counts.columns = ["시군구", "공영주차장_수"]

target_regions = [
    "전라남도 여수시", "경상북도 경주시", "전라북도 전주시",
    "경기도 가평군", "충청북도 단양군", "충청남도 보령시"
]

parking_counts[parking_counts["시군구"].isin(target_regions)]
```

```python
is_jeonju = (
    public["소재지지번주소"].fillna("").str.contains("전주시") |
    public["소재지도로명주소"].fillna("").str.contains("전주시")
)

jeonju_parking = public[is_jeonju]
print(f"전주시 공영주차장 개수: {len(jeonju_parking)}")
```

```python
jeonju_row = pd.DataFrame([{
    "시군구": "전라북도 전주시",
    "공영주차장_수": 110
}])

parking_counts_updated = pd.concat([parking_counts, jeonju_row], ignore_index=True)

parking_counts_updated.rename(columns={"시군구": "지역명"}, inplace=True)
target_regions = [
    "전라남도 여수시", "경상북도 경주시", "전라북도 전주시",
    "경기도 가평군", "충청북도 단양군", "충청남도 보령시"
]
result = parking_counts_updated[parking_counts_updated["지역명"].isin(target_regions)]
result
```

```python
# 접근성_점수 = (0.4 * IC_접근성_정규화 + 0.4 * 터미널_접근성_정규화 + 0.2 * 공영주차장_정규화)

df_merged = df_ic.merge(df_terminal, on="지역명")
df_merged = df_merged.merge(result.rename(columns={"지역명": "지역"}), left_on="지역명", right_on="지역")

df_merged["IC_접근성"] = 1 / df_merged["거리_km"]
df_merged["버스_접근성"] = 1 / df_merged["이동시간_분"]  
df_merged["공영주차장"] = df_merged["공영주차장_수"]

scaler = MinMaxScaler()
to_scale = ["IC_접근성", "버스_접근성", "공영주차장"]
scaled = scaler.fit_transform(df_merged[to_scale])
df_scaled = pd.DataFrame(scaled, columns=[col + "_정규화" for col in to_scale])

df_merged = pd.concat([df_merged, df_scaled], axis=1)

df_merged["접근성_점수"] = (
    0.4 * df_merged["IC_접근성_정규화"] +
    0.4 * df_merged["버스_접근성_정규화"] +
    0.2 * df_merged["공영주차장_정규화"]
)

final_score = df_merged[["지역명", "접근성_점수"]].sort_values(by="접근성_점수", ascending=False)
final_score
```

```python
import folium
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

trip_위치["주소"] = "전라남도 여수시 " + trip_위치["위치"]

geolocator = Nominatim(user_agent="yeosu_house_locator")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

latitudes = []
longitudes = []

for addr in empty_여수_sorted["주소"]:
    try:
        location = geocode(addr)
        if location:
            latitudes.append(location.latitude)
            longitudes.append(location.longitude)
        else:
            latitudes.append(None)
            longitudes.append(None)
    except:
        latitudes.append(None)
        longitudes.append(None)

empty_여수_sorted["위도"] = latitudes
empty_여수_sorted["경도"] = longitudes

trip_lat = []
trip_lng = []

for addr in trip_위치["주소"]:
    try:
        location = geocode(addr)
        if location:
            trip_lat.append(location.latitude)
            trip_lng.append(location.longitude)
        else:
            trip_lat.append(None)
            trip_lng.append(None)
    except:
        trip_lat.append(None)
        trip_lng.append(None)

trip_위치["위도"] = trip_lat
trip_위치["경도"] = trip_lng

m = folium.Map(location=[34.7604, 127.6622], zoom_start=11)

highlight = ["돌산읍", "화양면", "동문동", "화정면", "충무동"] # 빈집 수가 많은 상위 5개 지역 보라색으로 표시

for _, row in empty_여수_sorted.dropna(subset=["위도", "경도"]).iterrows():
    eupmyeondong = row["읍면동"]
    if eupmyeondong in highlight:
        folium.Marker(
            location=[row["위도"], row["경도"]],
            popup=f"<b>{row['주소']}<br>빈집수: {row['빈집수']}</b>",
            icon=folium.Icon(color='purple', icon='flag')
        ).add_to(m)
    else:
        folium.Marker(
            location=[row["위도"], row["경도"]],
            popup=f"{row['주소']}<br>빈집수: {row['빈집수']}",
            icon=folium.Icon(color='blue', icon='home')
        ).add_to(m)

for _, row in trip_위치.dropna(subset=["위도", "경도"]).iterrows():
    folium.Marker(
        location=[row["위도"], row["경도"]],
        popup=row["인기관광지"],
        icon=folium.Icon(color='red', icon='star')
    ).add_to(m)

m
```

> 같은 이유로 화양면과 화정면도 동일함
